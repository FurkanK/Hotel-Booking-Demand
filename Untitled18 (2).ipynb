```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
import time, datetime

X = df.drop('is_canceled', axis=1) # Öznitelikler
y = df.is_canceled # Etiketler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8,test_size=0.2, random_state=0)
```

```python
def fit_ml_algo(algo, X_train, y_train, cv):
    
    model = algo.fit(X_train, y_train)
    acc = round(model.score(X_train, y_train) * 100, 2)
    
    # Çapraz doğrulama(Cross Validation)
    train_pred = model_selection.cross_val_predict(algo, 
                                                  X_train, 
                                                  y_train, 
                                                  cv=cv, 
                                                  n_jobs = -1)
    acc_cv = round(metrics.accuracy_score(y_train, train_pred) * 100, 2)
    
    return train_pred, acc, acc_cv
```

```python
# Lojistik Regresyon
start_time = time.time()
train_pred_log, accuracy_log, acc_cv_log = fit_ml_algo(LogisticRegression(), 
                                                               X_train, 
                                                               y_train, 
                                                                    10)
log_time = (time.time() - start_time)
print("10-Katlamalı Cross Validation Doğruluğu: %s" % acc_cv_log)
print("Geçen Süre: %s" % datetime.timedelta(seconds=log_time))
```

```python
# Random Forest Algoritması
start_time = time.time()
train_pred_rf, acc_rf, acc_cv_rf = fit_ml_algo(RandomForestClassifier(), 
                                                                X_train, 
                                                                y_train,
                                                                10)
rf_time = (time.time() - start_time)
print("10-Katlamalı Cross Validation Doğruluğu: %s" % acc_cv_rf)
print("Geçen Süre: %s" % datetime.timedelta(seconds=rf_time))
```

```python
# Karar Ağacı Algoritması
start_time = time.time()
train_pred_dt, acc_dt, acc_cv_dt = fit_ml_algo(DecisionTreeClassifier(), 
                                                                X_train, 
                                                                y_train,
                                                                10)
dt_time = (time.time() - start_time)
print("10-Katlamalı Cross Validation Doğruluğu: %s" % acc_cv_dt)
print("Geçen Süre: %s" % datetime.timedelta(seconds=dt_time))
```

```python
# k-En yakın Komşuluk
start_time = time.time()
train_pred_knn, acc_knn, acc_cv_knn = fit_ml_algo(KNeighborsClassifier(), 
                                                  X_train, 
                                                  y_train, 
                                                  10)
knn_time = (time.time() - start_time)
print("10-Katlamalı Cross Validation Doğruluğu: %s" % acc_cv_knn)
print("Geçen Süre: %s" % datetime.timedelta(seconds=knn_time))
```

```python
# Lineer SVC
start_time = time.time()
train_pred_svc, acc_linear_svc, acc_cv_linear_svc = fit_ml_algo(LinearSVC(),
                                                                X_train, 
                                                                y_train, 
                                                                10)
svc_time = (time.time() - start_time)
print("10-Katlamalı Cross Validation Doğruluğu: %s" % acc_cv_linear_svc)
print("Geçen Süre: %s" % datetime.timedelta(seconds=svc_time))
```
